```python
cell_78=df[df['CellNo']==78]
cell_78=cell_78['ResistValue'].reset_index(drop=True)
cell_78.plot()
cell_78
```

```python
diff=[]
for x in range(len(cell_78)-1):
    diff.append(abs(cell_78[x]-cell_78[x+1]))
avg_diff = sum(diff)/(len(cell_78)-1)
print(avg_diff)
for i in range(len(cell_78)-1):
    if abs(cell_78[i]-cell_78[i+1])> avg_diff * 10 :
        print(i, end=' ')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

48번 셀, 누액 조치

```python
df=pd.read_csv('BANKS/m2_s2_bank_7ec19b22223bd644a1f18d17d90537dd__site_40ab7c891c559b478e96f8bb5c682ca0.csv')
tmp=df['TempValue']
for x in range(len(tmp)):
    if tmp[x]==-20:
        tmp[x]=np.nan
filled_df = tmp.interpolate()
df['TempValue']=filled_df
tmp=df[df['CellNo']==24].reset_index(drop=True)
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.plot(tmp['ResistValue'])
plt.subplot(1,3,2)
plt.plot(tmp['TempValue'])
plt.subplot(1,3,3)
plt.plot(tmp['VoltValue'])

tmp=df[df['CellNo']==26].reset_index(drop=True)
plt.figure(figsize=(15,5))
plt.subplot(2,3,1)
plt.plot(tmp['ResistValue'])
plt.subplot(2,3,2)
plt.plot(tmp['TempValue'])
plt.subplot(2,3,3)
plt.plot(tmp['VoltValue'])

# 얘는 정상
tmp=df[df['CellNo']==34].reset_index(drop=True)
plt.figure(figsize=(15,5))
plt.subplot(3,3,1)
plt.plot(tmp['ResistValue'])
plt.subplot(3,3,2)
plt.plot(tmp['TempValue'])
plt.subplot(3,3,3)
plt.plot(tmp['VoltValue'])
```
